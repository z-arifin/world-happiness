```python
rows_list = []
for region in data['Region'].unique():
    happinessRankSum = data.loc[data['Region'] == region]["Happiness Rank"].sum()
    numCountriesInRegion = len(data.loc[data['Region'] == region])
    avgHappinessScorePerRegion = happinessRankSum/numCountriesInRegion;
    dict1 = {'Region' : region, 'HappinessRankPerRegion' : avgHappinessScorePerRegion, 'numCountriesInRegion' : numCountriesInRegion};
    rows_list.append(dict1)
```

```python
df = pd.DataFrame(rows_list)
df.sort_values(["HappinessRankPerRegion"], ascending=True, inplace=True)
df
```

Based on the chart above, we see that Australia and New Zealand have the best Happiness Score followed closely by North America. However, note that both these regions have only 2 countries.

```python
plt.figure(figsize=(10, 10))
ax = sns.barplot(df['Region'], df['HappinessRankPerRegion'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```
